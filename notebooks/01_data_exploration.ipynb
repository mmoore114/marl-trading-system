```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Set plotting style
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
# Define the path to our raw data
data_path = Path("../data/raw/AAPL.csv")

# Load the data, using the FIRST column as the date index
df = pd.read_csv(data_path, index_col=0, parse_dates=True)

# List of columns to convert to a numeric type
cols_to_convert = ['Open', 'High', 'Low', 'Close', 'Volume']

# Loop through the columns and convert them
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# --- Verification ---
# Check the new data types
print("New data types:")
print(df.dtypes)

# Display the first 5 rows to see the cleaned data
df.head()
```

```python
# Plot the 'Close' price
df['Close'].plot(figsize=(15, 7), title='AAPL Closing Price History')
plt.ylabel('Price (USD)')
plt.show()
```
